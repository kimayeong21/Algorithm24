```python
def factrial_iter(n):       #축소 정복 기법: 상향식 접근
    result = 1
    for k in range(1,n+1):
        result = result * k #반복구조 이용
    return result 
```

```python
def factrial_iter(n):       
    result = 1
    for k in range(1,n+1):
        result = result * k 
    return result
n = int(input("자연수 입력: ")) 
print(n,"의 팩토리얼:",factrial_iter(n))
```
